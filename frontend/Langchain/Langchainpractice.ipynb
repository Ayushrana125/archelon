```python
# ================================
# AREX – DOCUMENT INGESTION LAB
# SHOW ALL CHUNKS (DEBUG MODE)
# ================================

import os
from pathlib import Path

from langchain_community.document_loaders import (
    TextLoader,
    PyPDFLoader,
    Docx2txtLoader
)

from langchain_text_splitters import RecursiveCharacterTextSplitter

print("🚀 Starting Arex ingestion test (FULL CHUNK VIEW)...\n")

# -------- CONFIG --------
DATA_FOLDER = r"D:\\-- Ayush Rana --\\Arex - Ayush Rana's Personal Assistant\\Langchain\\Data"
CHUNK_SIZE = 500
CHUNK_OVERLAP = 100
# ------------------------

# Step 1: Locate files
data_path = Path(DATA_FOLDER)
files = list(data_path.glob("*"))

print(f"📂 Found {len(files)} files:")
for f in files:
    print(f"   - {f.name}")
print()

# Step 2: Load documents
documents = []
print("📖 Loading documents...\n")

for file in files:
    try:
        if file.suffix.lower() == ".txt":
            loader = TextLoader(str(file), encoding="utf-8")
        elif file.suffix.lower() == ".pdf":
            loader = PyPDFLoader(str(file))
        elif file.suffix.lower() == ".docx":
            loader = Docx2txtLoader(str(file))
        else:
            print(f"⚠️ Skipping unsupported file: {file.name}")
            continue

        docs = loader.load()
        documents.extend(docs)

        print(f"✅ Loaded {file.name} → {len(docs)} sections")

    except Exception as e:
        print(f"❌ Error loading {file.name}: {e}")

print(f"\n📚 Total document sections: {len(documents)}\n")

# Step 3: Chunking
print("✂️ Chunking documents...\n")

splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)

chunks = splitter.split_documents(documents)

print(f"🧩 Total chunks created: {len(chunks)}\n")

# Step 4: SHOW ALL CHUNKS
print("🔎 DISPLAYING ALL CHUNKS\n")
print("=" * 80)

for idx, chunk in enumerate(chunks, start=1):
    source = chunk.metadata.get("source", "unknown")

    print(f"\n🧩 CHUNK #{idx}")
    print(f"📄 Source file: {os.path.basename(source)}")
    print(f"📏 Length: {len(chunk.page_content)} characters")
    print("-" * 80)
    print(chunk.page_content)
    print("-" * 80)

print("\n✅ All chunks displayed successfully.")
```

```python
# ================================
# AREX – DOCUMENT INGESTION LAB
# PART 1 (FINAL): LOAD + CHUNK + ID
# ================================

import os
import hashlib
from pathlib import Path
from collections import defaultdict

from langchain_community.document_loaders import (
    TextLoader,
    PyPDFLoader,
    Docx2txtLoader
)
from langchain_text_splitters import RecursiveCharacterTextSplitter

#print("🚀 Starting Arex ingestion test (PART 1 – final integrated)...\n")

# -------- CONFIG --------
DATA_FOLDER = r"D:\-- Ayush Rana --\Arex - Ayush Rana's Personal Assistant\Langchain\Data"
CHUNK_SIZE = 500
CHUNK_OVERLAP = 100
# ------------------------

# Utility: stable document_id from file path
def generate_document_id(path: str) -> str:
    return hashlib.md5(path.encode("utf-8")).hexdigest()

# Step 1: Locate files
data_path = Path(DATA_FOLDER)
files = list(data_path.glob("*"))

print(f"📂 Found {len(files)} files:")
for f in files:
    print(f"   - {f.name}")
print()

# Step 2: Load documents
documents = []
#print("📖 Loading documents...\n")

for file in files:
    try:
        if file.suffix.lower() == ".txt":
            loader = TextLoader(str(file), encoding="utf-8")
        elif file.suffix.lower() == ".pdf":
            loader = PyPDFLoader(str(file))
        elif file.suffix.lower() == ".docx":
            loader = Docx2txtLoader(str(file))
        else:
            print(f"⚠️ Skipping unsupported file: {file.name}")
            continue

        docs = loader.load()

        # attach document_id at load time
        document_id = generate_document_id(str(file.resolve()))
        for d in docs:
            d.metadata["document_id"] = document_id

        documents.extend(docs)
        print(f"✅ Loaded {file.name} → {len(docs)} sections | document_id={document_id}")

    except Exception as e:
        print(f"❌ Error loading {file.name}: {e}")

#print(f"\n📚 Total document sections loaded: {len(documents)}\n")

# Step 3: Chunking
#print("✂️ Chunking documents...\n")

splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)

chunks = splitter.split_documents(documents)

#print(f"🧩 Total chunks created: {len(chunks)}\n")

# Step 4: Assign chunk_index per document_id
#print("🔢 Assigning chunk_index per document...\n")

chunk_counters = defaultdict(int)

for chunk in chunks:
    doc_id = chunk.metadata["document_id"]
    chunk.metadata["chunk_index"] = chunk_counters[doc_id]
    chunk_counters[doc_id] += 1

#print("✅ chunk_index assigned.\n")

# Step 5: RAW MEMORY VIEW (ONE LINE PER CHUNK)
#print("🔎 RAW CHUNKS AS STORED IN MEMORY\n")
#print("=" * 140)

for chunk in chunks:
    print(
        f"{chunk.metadata} | {repr(chunk.page_content)}"
    )

print("\n✅ PART 1 COMPLETE: chunks are uniquely identifiable, ordered, and embedding-ready.")
```

```python
# ================================
# AREX – DOCUMENT INGESTION LAB
# PART 2: EMBEDDINGS (MISTRAL)
# ================================

import os
from langchain_mistralai import MistralAIEmbeddings

# --- CONFIG ---
#MISTRAL_API_KEY = os.getenv("MISTRAL_API_KEY")  # or set directly (not recommended)
MISTRAL_API_KEY = "JZfkCrdJuZrxg7M9Y3AlIl4DLujRMaMR"

EMBED_MODEL = "codestral-embed"
# --------------

print("🧠 Starting embedding generation...\n")

# Step 1: Initialize embedding model
embedding_model = MistralAIEmbeddings(
    api_key=MISTRAL_API_KEY,
    model=EMBED_MODEL
)

# Step 2: Prepare raw texts (EXACTLY what will be embedded)
texts_to_embed = [chunk.page_content for chunk in chunks]

print(f"📦 Total texts to embed: {len(texts_to_embed)}\n")

# Step 3: Generate embeddings
embeddings = embedding_model.embed_documents(texts_to_embed)

print("✅ Embeddings generated.\n")

# Step 4: RAW OUTPUT (what actually happened)
print("🔎 RAW EMBEDDING OUTPUT\n")
print("=" * 120)

for i, (chunk, vector) in enumerate(zip(chunks, embeddings)):
    print(
        f"document_id={chunk.metadata['document_id']} | "
        f"chunk_index={chunk.metadata['chunk_index']} | "
        f"vector_dim={len(vector)} | "
        f"vector_sample={vector}"
    )

print("\n✅ PART 2 COMPLETE: embeddings ready for vector DB.")
```

```python
# ================================
# AREX – DOCUMENT INGESTION LAB
# PART 3: INSERT INTO SUPABASE
# ================================

from dotenv import load_dotenv
load_dotenv()

from supabase import create_client

# --- CONFIG ---
SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_SERVICE_ROLE_KEY")  # service role, not anon
TABLE_NAME = "documents_chunks"
# ----------------

print("📤 Inserting embeddings into Supabase...\n")

# Step 1: Create Supabase client
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

# Step 2: Build rows to insert
rows = []

for chunk, vector in zip(chunks, embeddings):
    row = {
        "document_id": chunk.metadata["document_id"],
        "chunk_index": chunk.metadata["chunk_index"],
        "content": chunk.page_content,
        "embedding": vector,
        "metadata": {
            k: v for k, v in chunk.metadata.items()
            if k not in ["document_id", "chunk_index"]
        }
    }
    rows.append(row)

print(f"📦 Prepared {len(rows)} rows for insertion\n")

# Step 3: Insert into Supabase
response = supabase.table(TABLE_NAME).insert(rows).execute()

# Step 4: Raw response
print("🔎 Supabase response:\n")
print(response)

print("\n✅ PART 3 COMPLETE: embeddings stored in Supabase.")
```
